# 🛡️ Exploratory Data Analysis (EDA) on Security Datasets

This notebook explores the baseline CICIDS2017 / UNSW-NB15 network traffic datasets to characterize benign vs. malicious attributes, feature distributions, and protocol frequencies.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from data_utils import generate_synthetic_data

# Load synthetic baseline dataset representing network logs
df = generate_synthetic_data(n_samples=5000, attack_ratio=0.15)
print(f"Dataset shape: {df.shape}")
df.head()
```

```python
# Analyze label distributions
print(df['label'].value_counts())
sns.countplot(data=df, x='label')
plt.title('Distribution of Traffic Labels')
plt.xticks(rotation=45)
plt.show()
```
